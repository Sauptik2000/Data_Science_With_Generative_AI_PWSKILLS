#Functions

**1. What is the difference between a function and a method in Python?**
- In Python, both functions and methods are callable, but there are key differences:
  - Function:
      * A function is a block of code that is defined using the def keyword and can be called independently.
      * It can be defined outside of any class.
      * Functions are called by their name and can accept parameters, perform operations, and return a value.
  - Method:
    * A method is a function that is associated with an object or an instance of a class.
    * Methods are defined within a class and are called using the object or class.
    * The first parameter of a method is usually self, which refers to the instance of the class.
- Key Difference:
  * Function: Defined independently, not tied to an object or class.
  * Method: Defined inside a class and tied to an object or class instance.

**2. Explain the concept of function arguments and parameters in Python.**
- **Parameters** are variables defined in the function definition that act as placeholders for the values the function will receive.
  
  Example:
  ```python
  def greet(name):  # 'name' is the parameter
      return f"Hello, {name}!"
  ```

- **Arguments** are the actual values passed to the function when calling it.
  
  Example:
  ```python
  greet("Alice")  # "Alice" is the argument
  ```

- Key Points:
  - **Parameters**: Defined in the function definition.
  - **Arguments**: Passed when calling the function.

**3. What are the different ways to define and call a function in Python?**
- In Python, functions can be defined and called in various ways:
1. **Defining a Function**:Using the `def` keyword:
```python
def greet(name):
    return f"Hello, {name}!"
```
2. **Calling a Function**:
Call by using the function name followed by parentheses:
```python
greet("Alice")
```

3. **Default Arguments**:
Functions can have default values for parameters:
```python
def greet(name, age=25):
    return f"Hello, {name}. You are {age} years old."
```

4. **Variable-Length Arguments**:
Use `*args` or `**kwargs` to pass a variable number of arguments:
```python
def add(*numbers):
    return sum(numbers)
```

5. **Lambda Functions**:
Anonymous functions defined with `lambda`:
```python
square = lambda x: x ** 2
```

- Summary:
  - **Define** using `def`.
  - **Call** by function name with parentheses.
  - **Use default or variable-length arguments**.
  - **Lambda** for small, unnamed functions.

**4. What is the purpose of the `return` statement in a Python function?**
- The `return` statement in a Python function is used to send a result or value back to the caller, effectively ending the function's execution and passing the result out.

- Purpose of `return`:
  1. **Returns a value**: Sends the result of the function to the caller.
  2. **Exits the function**: Stops further execution of the function after the `return` statement.

- Example:
```python
def add(a, b):
    return a + b  # Returns the sum of a and b
result = add(3, 5)  # result will be 8
```

- Summary: The `return` statement provides the output of the function and terminates its execution.

**5.What are iterators in Python and how do they differ from iterables?**
- In Python, **iterators** and **iterables** are closely related but have distinct roles:

1. **Iterable**:
- An **iterable** is any Python object capable of returning its members one at a time, allowing it to be looped over (e.g., using a `for` loop).
- It defines an `__iter__()` method that returns an **iterator**.

- Examples of Iterables: Lists, tuples, strings, dictionaries, sets, etc.
```python
my_list = [1, 2, 3]
```

2. **Iterator**:
- An **iterator** is an object that represents a stream of data, and it is used to iterate over an iterable.
- It defines two methods:
  - `__iter__()` (returns the iterator itself)
  - `__next__()` (returns the next item from the iterable)

- Example of an Iterator:
```python
my_list = [1, 2, 3]
iterator = iter(my_list)  # Converts the iterable into an iterator
print(next(iterator))  # 1
print(next(iterator))  # 2
```
Key Differences:
- **Iterable**: Can be any object that can return an iterator (e.g., a list).
- **Iterator**: The object that is actually used to traverse the iterable (via `next()`).


**6. Explain the concept of generators in Python and how they are defined**
- In Python, **generators** are a type of iterable, like lists or tuples, but they are more memory efficient because they generate values on the fly rather than storing them all in memory at once. They are defined using a function with the `yield` keyword instead of `return`.

- Key Points:
1. **yield**: Instead of returning a single value, the `yield` keyword "pauses" the function, saving its state, and can be resumed later to generate the next value.
2. **Lazy Evaluation**: Generators only produce values when requested, which helps in handling large datasets or infinite sequences efficiently.

- Example of a simple generator:
```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
gen = count_up_to(5)
for number in gen:
    print(number)   # Output: 1 2 3 4 5
```
In this case, the generator `count_up_to` yields numbers one by one, and once the function finishes, the generator is exhausted.

**7. What are the advantages of using generators over regular functions?**
- Generators are more memory-efficient because they yield one value at a time, rather than storing all values in memory. They use **lazy evaluation**, meaning they only generate values when requested, which is particularly useful for working with large datasets or infinite sequences. This leads to **better performance** and **faster execution**, as it avoids the overhead of processing or storing unnecessary data. In contrast, regular functions return all values at once, which can consume more memory and computational resources.

**8. What is a lambda function in Python and when is it typically used?**
- A **lambda function** in Python is a small, anonymous function defined using the `lambda` keyword. It can take any number of arguments but can only have one expression. The result of the expression is automatically returned.

- Syntax:
```python
lambda arguments: expression
```
- Example:
```python
add = lambda x, y: x + y
print(add(2, 3))  # Output: 5
```
- When to use:
  - **Short, throwaway functions**: When we need a simple function for a short period, like in sorting or filtering data.
  - **Functional programming**: Typically used with functions like `map()`, `filter()`, or `reduce()` to perform operations without defining a full function.
  
In general, lambda functions are used when we need a small function for a specific task, without the overhead of defining a full `def` function.

**9. Explain the purpose and usage of the `map()` function in Python**
- The **`map()`** function in Python is used to apply a given function to each item in an iterable (like a list, tuple, etc.) and return a map object (which is an iterator) that contains the results.

- Syntax:
```python
map(function, iterable)
```
- Purpose:
  - It allows you to **transform data** efficiently by applying a function to each element in an iterable.

- Example:
```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```
- Usage:
  - Commonly used for **transforming or processing data** in an iterable without needing an explicit loop. It can be combined with **lambda functions** for concise code.



**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**
- Here’s a brief comparison of `map()`, `reduce()`, and `filter()` in Python:
1. **`map()`**:
   - **Purpose**: Applies a given function to each item in an iterable and returns a new iterable with the results.
   - **Usage**: Transforming data.
   - **Example**: Squaring each number in a list.
   ```python
   numbers = [1, 2, 3]
   squared = map(lambda x: x**2, numbers)
   print(list(squared))  # Output: [1, 4, 9]
   ```
2. **`reduce()`**:
   - **Purpose**: Applies a function cumulatively to the items of an iterable, reducing it to a single value.
   - **Usage**: Aggregating or accumulating values (e.g., summing a list).
   - **Example**:
   ```python
   from functools import reduce
   numbers = [1, 2, 3]
   result = reduce(lambda x, y: x + y, numbers)
   print(result)  # Output: 6
   ```
3. **`filter()`**:
   - **Purpose**: Filters elements from an iterable based on a condition, keeping only the items for which the function returns `True`.
   - **Usage**: Filtering data.
   - **Example**: Keeping only even numbers in a list.
   ```python
   numbers = [1, 2, 3, 4]
   even_numbers = filter(lambda x: x % 2 == 0, numbers)
   print(list(even_numbers))  # Output: [2, 4]
   ```
- Summary:
  - **`map()`**: Transforms each item.
  - **`reduce()`**: Aggregates all items into a single result.
  - **`filter()`**: Filters items based on a condition.

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n):
        self.n, self.a, self.b = n, 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.n == 0:
            raise StopIteration
        self.n -= 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

for num in FibonacciIterator(10):
    print(num)
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_2(5):
    print(power)
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_name):
    with open(file_name, 'r') as file:
        for line in file:
            yield line.strip()
````

````python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

input_string = "Hello, World!"

def remove_vowels(char):
    return char.lower() not in 'aeiou'

filtered_string = ''.join(filter(remove_vowels, input_string))

print(filtered_string)
````
